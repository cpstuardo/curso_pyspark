```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

### Función reduce

Aplica la función de reducción de todos los elementos del RDD y la envía al drive

```python
# Suma de todos los elementos del RDD
rdd = sc.parallelize([2,4,6,8])
rdd.reduce(lambda x,y: x + y)
```

```python
# multiplicación de los elementos del RDD
rdd1 = sc.parallelize([1,2,3,4])
rdd1.reduce(lambda x,y: x * y)
```

### Función count

Cuenta el número de elementos del RDD y lo envía al driver

```python
rdd = sc.parallelize(['j', 'o', 's', 'e'])
rdd.count()
```

```python
rdd1 = sc.parallelize([item for item in range(10)])
rdd1.count()
```

### Función collect

Recopila los elemetos del RDD (de todas las particiones) y las envía al driver

```python
rdd = sc.parallelize('Hola Apache Spark!'.split(' '))
rdd.collect()
```

```python
rdd1 = sc.parallelize([(item, item ** 2) for item in range(20)])
rdd1.collect()
```

### Funciónes take, max y saveAsTextFile

```python
# take: retorna los primeros n elementos del RDD
rdd = sc.parallelize('La programación es bella'.split(' '))
```

```python
rdd.take(2)
```

```python
rdd.take(4)
```

```python
# max: entrega el elemento máximo del RDD
rdd1 = sc.parallelize([item/(item + 1) for item in range(10)])
rdd1.max()
```

```python
# saveAsTextFile: 
rdd.collect()
rdd.saveAsTextFile('./rdd')
```

```python
# Para que cree sólo un fichero, reduce las particiones primero
rdd.coalesce(1).saveAsTextFile('./rdd1')
```
